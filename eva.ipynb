```python
import pandas as pd
import numpy as np
```

```python
df_test1 = pd.read_csv("test1/test.tsv",sep="\t")
df_test2 = pd.read_csv("test2/test.tsv",sep="\t")
```

```python
res1 = pd.read_csv("res1/test_results.tsv",sep="\t",header=None)
res2 = pd.read_csv("res2/test_results.tsv",sep="\t",header=None)
```

```python
df_eva = df_test1[["InputStoryid","AnswerRightEnding"]]
```

```python
df_eva["score1"] = res1[1]
df_eva["score2"] = res2[1]
```

```python
scores = df_eva[["score1","score2"]].values
pred = np.argmax(scores,1)+1
df_eva["pred"] = pred
```

```python
np.mean(df_eva["pred"].values == df_eva["AnswerRightEnding"].values)
```
